Напишіть функції, що:
* рохраховує мінімальне та максимальне значення
* середнє значення послідовності
* медіану послідовності
* частоту кожної цифри в послідовності
* середнє квадратичне відхилення

Послідовність чисел ганеруємо рандомно

```python
import random

#генеруэмо тисячу чисел в діапазоні від 0 до 10
list_with_random_numbers = [random.randint(0,100) for i in range(1000)]

max_number = max(list_with_random_numbers)
print("Максимальне число: " , max_number)

min_number = min(list_with_random_numbers)
print("Мінімальне число: " , min_number)


average_value = sum(list_with_random_numbers) / 1000
print("Середнє значення: ", average_value)

#медіана
sorted_list = sorted(list_with_random_numbers)
mediana = (sorted_list[500] + sorted_list[501]) / 2
print("Медіана: ", mediana)

Dictionary = {}
for i in list_with_random_numbers:
    Dictionary[i] = list_with_random_numbers.count(i)
print("Частота кожної цифри в послідовності: " , Dictionary)

#мода
max_count = max(Dictionary.values())
most_frequent = [k for k, v in Dictionary.items() if v == max_count]
print("Мода: ", most_frequent)



                    
```
